```python
environment_df = pd.read_csv("appearance_living_environment.csv", encoding='latin-1')
environment_df
```

```python
len(environment_df.Name.unique())
```

```python
import numpy as np
import matplotlib.pyplot as plt  # To visualize
import pandas as pd  # To read data
from sklearn.linear_model import LinearRegression
```

```python
environment_df_model = environment_df[environment_df.isna().any(axis=1) == False]

X = environment_df_model[["Year","Name"]]
X = pd.get_dummies(data=X, drop_first=True)
Y = environment_df_model["Appearance living environment (1-10)"]
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
    
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .20, random_state = 40)
    
regr = linear_model.LinearRegression() # Do not use fit_intercept = False if you have removed 1 column after dummy encoding
regr.fit(X_train, Y_train)
predicted = regr.predict(X_test)
```

```python
locations = environment_df["Name"].unique()

environment_2023 = pd.DataFrame({
    'Year': [2023] * len(locations),
    'Name': locations
})

# Step 2: Apply one-hot encoding
environment_2023_encoded = pd.get_dummies(data=environment_2023, drop_first=True)

# Step 1: Reorder columns to match the order in the training data
environment_2023_encoded = environment_2023_encoded[X_train.columns]

# Step 2: Remove any additional columns not present in the training data
environment_2023_encoded = environment_2023_encoded[X_train.columns.intersection(environment_2023_encoded.columns)]

# Step 3: Use the trained model for prediction
predicted_2023 = regr.predict(environment_2023_encoded)

# Add the predicted values to the DataFrame
environment_2023['Predicted appearance living environment (1-10)'] = predicted_2023

print(environment_2023)
```

```python
environment_2023['Predicted appearance living environment (1-10)'] = environment_2023['Predicted appearance living environment (1-10)'].apply(lambda x: round(x, 1))
environment_2023
```

```python
areas_df = pd.read_csv("areas.csv", encoding='latin-1')
areas_df = areas_df[areas_df['Year'] == 2022]
# the areas haven't changed so will use latest data from 2022
areas_df = areas_df.drop(['Year'], axis=1)
# convert to square kilometers
areas_df['Area total (ha)']= areas_df['Area total (ha)']/100
areas_df.rename(columns={"Area total (ha)": "Area total (km2)"}, inplace=True)
areas_df
```

```python
income_df = pd.read_csv("mean_disposable_income.csv", encoding='latin-1')
income_df
```

```python
len(income_df.Name.unique())
```

```python
income_df_model = income_df[income_df.isna().any(axis=1) == False]

X = income_df_model[["Year","Name"]]
X = pd.get_dummies(data=X, drop_first=True)
Y = income_df_model["Mean disposable household income"]
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
    
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .20, random_state = 40)
    
regr = linear_model.LinearRegression() # Do not use fit_intercept = False if you have removed 1 column after dummy encoding
regr.fit(X_train, Y_train)
predicted = regr.predict(X_test)
```

```python
locations = environment_df["Name"].unique()

income_2023 = pd.DataFrame({
    'Year': [2023] * len(locations),
    'Name': locations
})

# Step 2: Apply one-hot encoding
income_2023_encoded = pd.get_dummies(data=income_2023, drop_first=True)

# Step 1: Reorder columns to match the order in the training data
income_2023_encoded = income_2023_encoded[X_train.columns]

# Step 2: Remove any additional columns not present in the training data
income_2023_encoded = income_2023_encoded[X_train.columns.intersection(income_2023_encoded.columns)]

# Step 3: Use the trained model for prediction
predicted_2023 = regr.predict(income_2023_encoded)

# Add the predicted values to the DataFrame
income_2023['Predicted mean disposable household income'] = predicted_2023

print(income_2023)
```

```python
income_2023['Predicted mean disposable household income'] = income_2023['Predicted mean disposable household income'].apply(lambda x: round(x, 1))
income_2023
```

```python
# merge all dataframes together
all_data = [df_merged, income_2023, environment_2023]
all_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Name', 'Year'],
                                             how='outer'), all_data)
all_merged
```
